```python
from random import choice
from itertools import count
from seaborn import histplot
```

```python
options = [j for j in range(9)]
```

```python
N = 100000
streaks = []
minimal_streaks = []
```

```python
for k in range(N):
    j = 0
    games = []
    for i in count():
        game = choice(options)
        games.append(game)
        if game == options[j]:
            j += 1
        if j == len(options):
            break
    streaks.append(i)
    j = len(options) - 1
    for i, game in enumerate(reversed(games)):
        if game == options[j]:
            j -= 1
        if j == 0:
            break
    minimal_streaks.append(i)
```

```python
print(min(streaks), sum(streaks)//len(streaks), max(streaks), sorted(streaks)[len(streaks) // 2])
print(min(minimal_streaks), sum(minimal_streaks)//len(minimal_streaks), max(minimal_streaks), sorted(minimal_streaks)[len(streaks) // 2])
```

```python
histplot(streaks, binwidth=1)
```

```python
histplot(minimal_streaks, binwidth=1)
```

```python
from collections import defaultdict

streak_lengths = defaultdict(int)
for i in streaks:
    streak_lengths[i] += 1
```

```python
print(sorted(streak_lengths.items()))
```
